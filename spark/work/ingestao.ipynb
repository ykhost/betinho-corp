```python
from pyspark.sql import SparkSession
import requests
spark = SparkSession.builder.appName('load').enableHiveSupport().getOrCreate()
```

```python
def loadData (qtde):
    list = []
    for x in range (qtde):
        print(x)
        r = requests.get('https://random-data-api.com/api/v2/users')
        list.append(r.json())
        req = spark.read.json(spark.sparkContext.parallelize(list))
        req = req.select( \
         'email' \
        ,'first_name' \
        ,'last_name' \
        ,'gender' \
        ,'id' \
        ,'username' \
                 )
    return req
```

```python
df = loadData(10)
```

```python
df.write.json('/datalake/raw/user',mode='append')
```

```python
for x in range (20):
    df = loadData(10)
    df.write.json('/datalake/raw/user',mode='append')
```
